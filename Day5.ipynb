## Violin Plots

```python
N = 10**4
np.random.seed(42)
```

```python
def plot_comparison(x, title):
    fig, ax = plt.subplots(3, 1, sharex=True)
    sns.kdeplot(x, ax=ax[0])
    ax[0].set_title('Histogram + KDE')
    sns.boxplot(x, ax=ax[1])
    ax[1].set_title('Boxplot')
    sns.violinplot(x, ax=ax[2])
    ax[2].set_title('Violin plot')
    fig.suptitle(title, fontsize=16)
    plt.show()
```

```python
plot_comparison(np.random.normal(size=N) , 'Normal Distribution')
```

```python
plot_comparison(np.random.lognormal(size=N) , 'Log Normal Distribution')
```

```python
tips = sns.load_dataset("tips")

ax = sns.violinplot(x="sex", y="tip", inner='quartile', data=tips)
ax.set_title('Distribution of tips', fontsize=16)
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
ax.set_title('Distribution of total bill amount per day', fontsize=16)
```

conclusion : 
Wider sections of the violin plot represent a higher probability that members of the population will take on the given value; the skinnier sections represent a lower probability

### defaultdict

```python
from collections import defaultdict

ordinary_dict = dict()
default_dict = defaultdict(list)

# accessing not found keys 

print(ordinary_dict.get('fashion',[]))  #lazem t-state el none

default_dict['fashion'] #msh lazem
```

### ChainMap : 
ChainMap is used to combine multiple dictionaries or mappings into a single view, allowing us to search and access keys in all the dictionaries as if they were a single dictionary

```python
from collections import ChainMap

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
combined_dict = ChainMap(dict1, dict2)
print(combined_dict['a'])  # Output: 1
print(combined_dict['c'])  # Output: 3


# replace with 
recombined_dict = {**dict1 , **dict2}
print(recombined_dict['a'])
print(recombined_dict['c'])
```

### flattened matrix in one-line

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_matrix = [num for row in matrix for num in row]
print(flattened_matrix)  
# Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```
