### with columns as group by 

```python
# with columns as ""15- Pandas_Chain_GroupBy_TS.ipynb"
df = pd.DataFrame({
    "class": ["a", "a", "a", "b", "b", "b", "b"],
    "score": ["80", "39", "67", "28", "77", "90", "44"],
})

df["score"] = df["score"].astype(int)

df['avg_score']  = df.groupby("class")["score"].transform("mean")


```

```python
df = pl.DataFrame({
    "class": ["a", "a", "a", "b", "b", "b", "b"],
    "score": [80, 39, 67, 28, 77, 90, 44],
})

df.with_columns(
    pl.col("score").mean().over("class").alias("avg_score")
)
```

### Piping 

```python
pd_df.rename(columns={"Shipping_Cost(USD)": "price"},inplace=True)
pd_df.head()
```

```python
def discount(df):
    df["30_percent_off"] = df["price"] * 0.7
    return df
def vat(df):
    df["vat"] = df["price"] * 0.14
    return df
def total_cost(df):
    df["total"] = df["30_percent_off"] + df["vat"]
    return df
(pd_df
 .pipe(discount)
 .pipe(vat)
 .pipe(total_cost)
)

```

```python
def discount(input_col):
    return pl.col(input_col).mul(0.7).alias("30_percent_off")
def vat(input_col):
    return pl.col(input_col).mul(0.2).alias("vat")
def total_cost(input_col1, input_col2):
    return pl.col(input_col1).add(pl.col(input_col2).alias("total"))
                                  
pl_df.with_columns(
    discount("Shipping_Cost(USD)"),
    vat("Shipping_Cost(USD)"),
    total_cost("30_percent_off", "vat"),
)
```
