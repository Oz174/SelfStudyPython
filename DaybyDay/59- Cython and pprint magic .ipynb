```python
def foo(check):
    count = 0 
    for i in range(10_000):
        for j in range(10_000):
            if(i+j)%11 == check:
                count+=1
    return count
```

```python
%%cython -a
def foo_c(int check):
    cdef int i , j
    cdef int count = 0
    for i in range(10_000):
        for j in range(10_000):
            if(i+j)%11 == check:
                count+=1
    return count
```

```python
foo_c(2)
```

```python
foo(2)
```

```python
%timeit -r 1 -n 1 foo_c(2)
```

```python
%timeit -r 1 -n 1 foo(2)
```
