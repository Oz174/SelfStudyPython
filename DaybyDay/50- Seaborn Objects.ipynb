# Imports

```python
import seaborn.objects as so 
import seaborn as sns
import pandas as pd
```

# Topics

```python
# let's import sns tips dataset 

tips = sns.load_dataset('tips')
tips.head(3)
```

## Basic Syntax 

1. A call to Plot(),
2. Assignment of a data argument,
3. Assignment of a coordinate argument (such as x and/or y),
4. A call to the add() method to add a layer to the plot,
5. A call to a method inside add() to specify the marker/plot type.

```python
so.Plot()
```

```python
so.Plot(data = tips , x='total_bill',y='tip')
```

```python
so.Plot(data = tips , x='total_bill',y='tip').add(so.Dot()) # basic 
```

You could see other Plot() Methods beside add here in this image 
![image.png](./images/Plot_methods.webp)

```python
so.Plot(data = tips , x='total_bill',y='tip').add(so.Dot(), color='sex') # filtered
```

```python
# we could add a ... trendline 
so.Plot(data=tips, x='total_bill', y='tip') \
.add(so.Dot(),color='sex') \
.add(so.Line(color='red'),so.PolyFit()) \
.label(title="Total Bill Vs Tips")
```

```python
fig = (so.Plot(tips, 'total_bill', 'tip')
       .add(so.Dot(), color='sex')        
       .add(so.Line(color='red'), so.PolyFit()))

fig.facet("sex")
```

To add a “super” title across the top of a facet grid, you have to rely on Matplotlib’s pyplot module

```python
import matplotlib.pyplot as plt

fig = plt.Figure(figsize=(5, 6))
fig.suptitle("Tips vs. Total Bill by Gender")

(
    so.Plot(tips, 'total_bill', 'tip')
           .add(so.Dot(), color='sex')  
           .add(so.Line(color='red'), so.PolyFit())
           .facet(col='sex')
           .on(fig)
)   
```

```python
(
    so.Plot(tips, y='tip')
    .pair(x=['total_bill', 'size', 'sex'])
    .add(so.Dots(), color='sex')
)
```

```python
(
so.Plot(tips, x='sex')
    .add(so.Bar(), 
         so.Count(), 
         color='sex')
)
```

```python
# is there is a relation between the smoker and the tip ?
so.Plot(tips, y='tip') \
    .pair(x=['smoker','day']) \
    .add(so.Dots(), color='time')
    
```

```python
# who gives more tips ? smokers or non smokers ?
so.Plot(tips, x='smoker') \
    .add(so.Bar(), so.Count(), color='smoker') \
    .label(title="Smokers vs Non Smokers")
```
