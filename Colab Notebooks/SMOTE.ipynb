```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE

# Create a synthetic imbalanced dataset
data, target = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                                   n_clusters_per_class=1, weights=[0.9], flip_y=0,
                                   random_state=42)

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])
df['Target'] = target

# Visualize the original dataset
sns.scatterplot(data=df, x='Feature1', y='Feature2', hue='Target', alpha=0.5)
plt.title('Original Imbalanced Dataset')
plt.show()
```

```python
# Initialize SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)

# Fit and resample the dataset
X_resampled, y_resampled = smote.fit_resample(data, target)

# Convert to DataFrame for visualization
df_resampled = pd.DataFrame(X_resampled, columns=['Feature1', 'Feature2'])
df_resampled['Target'] = y_resampled

# Visualize the resampled dataset
sns.scatterplot(data=df_resampled, x='Feature1', y='Feature2', hue='Target', alpha=0.5)
plt.title('Dataset After Applying SMOTE')
plt.show()
```
