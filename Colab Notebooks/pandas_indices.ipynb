```python
import pyarrow
import pandas as pd
import numpy as np
```

```python
names = ['bob' , 'axel' , 'john' , 'adam' , 'smith' , 'isaac']
instruments = ['guitar' , 'piano' , 'drums' , 'flute' , 'violet']

df = pd.DataFrame(
    {
        'first_name' : np.random.choice(names , 100) ,
        'last_name' : np.random.choice(names , 100) ,
        'instrument' : np.random.choice(instruments , 100),
        'birth_date' : pd.date_range('1980-01-01' , '2000-01-01' , periods = 100)
    }
)
```

```python
df.head()
```

```python
df.sort_values('first_name').head()
```

Here what happens to the indices on sorting

```python
df.sort_values('first_name' , ignore_index = True).head()
```

```python
# equivalent to
df.sort_values('first_name').reset_index(drop = True).head()
```

```python
# equivalent to
df.sort_values('first_name').set_axis(range(100)).head()
```

```python
# what if drop is false ?
df.sort_values('first_name').reset_index(drop = False).head()
```

```python
df2 = df.sort_values('first_name').copy()
print(df2.head())
df2.index = pd.RangeIndex(start = 0 , stop = 100 , step = 1)
df2.head()
```

```python
df2.loc[[4]] #--> returns by index
```

```python
df2.iloc[[4]] #--> returns by position
```

## Fixing Duplicate Indices

```python
guitar_players = df2[df2['instrument'] == 'guitar']
guitar_players = guitar_players.reset_index(drop=True)
guitar_players.head()
```

```python
non_guitar_players = df2[df2['instrument'] != 'guitar']
non_guitar_players = non_guitar_players.reset_index(drop=True)
non_guitar_players.head()
```

```python
merged = pd.concat([guitar_players , non_guitar_players] , axis = 0)
```

```python
# we have duplicate indices , we can use loc but not iloc

merged.loc[0]
```

```python
try :
  print(merged.filter(items=[0,1] , axis="index"))
except Exception as e :
  print(e)
```

## Aligning Indices

you will have to reset the index of both or one of them

```python
week1_sales = pd.DataFrame(
  {
      'Day' : ['Mon' , 'Tue' , 'Wed' , 'Thu' , 'Fri'] ,
      'Sales' : [100 , 200 , 300 , 400 , 500]
  },
  index = range(1,6))

week1_sales
```

```python
week2_sales = week1_sales.reset_index(drop=True)
week2_sales
```

```python
week1_sales['Sales'] + week2_sales['Sales']
```

```python
# why did it happen

merged_df = week1_sales.merge(week2_sales , how = 'inner' , left_index=True , right_index=True)
merged_df
```

```python
merged_df = week1_sales.merge(week2_sales , how = 'outer' , left_index=True , right_index=True)
merged_df
```

```python
week1_sales = week1_sales.reset_index(drop=True)

week1_sales['Sales'] + week2_sales['Sales']
```
